```python
from networkx.generators.random_graphs import gnp_random_graph
from cvxpy import Maximize, Variable, trace, Problem, cumsum, sum

import matplotlib.pyplot as plt
%matplotlib inline
import networkx as nx
import numpy as np
import random
```

```python
n = 100
k = 30
# np.random.choice(range(100),30)
# cliqueNodes.sort()
cliqueNodes = [2, 4, 10, 14, 16, 17, 17, 18, 20, 22, 23, 27, 40, 48, 54, 58, 58, 60, 62, 71, 73, 73, 74, 76, 79, 81, 84, 86, 91, 96]
```

```python
G = gnp_random_graph(n, 0.3)
for i in cliqueNodes:
    for j in cliqueNodes:
        if i != j: G.add_edge(i,j)
```

```python
plt.figure(figsize=(15, 12))
nx.draw_networkx(G, width=0.5)
```

```python
X = Variable((n, n), PSD=True)
objective = Maximize(sum(X))
```

```python
constraints = []
for i,j in G.edges:
    constraints.append(X[i, j] == 0)
constraints.append(trace(X) == 1)
        
# Solve the program
problem = Problem(objective, constraints)
problem.solve()
```

```python
X_sol = problem.variables()[0].value
eigenvalues, eigenvectors = np.linalg.eig(X_sol)
```

```python
eigenvectors
```

```python
topeig = eigenvectors[:, 1]
```

```python
np.argmax(topeig)
```

```python
# copied from wilson's code
A = np.zeros((n, n))
for e in G.edges:
    A[e[0]][e[1]] = 1
    A[e[1]][e[0]] = 1

x = [abs(i) for i in topeig.tolist()]
largest = sorted(x, reverse = True)[:k]

I = [x.index(l) for l in largest]
I.sort()
print(I)
adj = np.array([1 if i in I else 0 for i in range(n)])

A_adj = (A @ adj).tolist()

found_clique = [i for i in range(n) if A_adj[i] >= 0.75 * k]
```
