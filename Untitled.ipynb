```python
from networkx.generators.random_graphs import gnp_random_graph
from cvxpy import Maximize, Variable, trace, Problem, cumsum, sum

import matplotlib.pyplot as plt
%matplotlib inline
import networkx as nx
import numpy as np
import random
```

```python
n = 300
k = 60
cliqueNodes = np.random.choice(range(n),k,replace=False)
cliqueNodes.sort()
print(cliqueNodes)
```

```python
G = gnp_random_graph(n, 0.3)
for i in cliqueNodes:
    for j in cliqueNodes:
        if i != j: G.add_edge(i,j)
```

```python
plt.figure(figsize=(15, 12))
nx.draw_networkx(G, width=0.5)
```

```python
X = Variable((n, n), PSD=True)
objective = Maximize(sum(X))
```

```python
constraints = []
for i,j in G.edges:
    constraints.append(X[i, j] == 0)
constraints.append(trace(X) == 1)
        
# Solve the program
problem = Problem(objective, constraints)
problem.solve()
```

```python
X_sol = problem.variables()[0].value
eigenvalues, eigenvectors = np.linalg.eig(X_sol)
```

```python
eigenvectors
```

```python
topeig = eigenvectors[:, 0]
```

```python
a = np.sort(topeig.argsort()[-(k):])
# print([i for i in range(100) if i not in a])
a
```

```python
print(cliqueNodes)
```

```python
# for i in range(len(a)):
#     for j in range(i+1, len(a)):
#         if (a[i],a[j]) not in G.edges:
#             print(a[i],a[j])
p,q = 0,0
count = 0
while p < k and q < k:
    if cliqueNodes[p] == a[q]:
        p += 1
        q += 1
    elif cliqueNodes[p] > a[q]:
        q += 1
        count += 1
    else:
        p += 1
        count += 1
print(count) 
# G.edges
```
